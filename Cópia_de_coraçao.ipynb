## **Transformando as variáveis categóricas nominais em variáveis categóricas ordinais** `




```python
df2 = pd.DataFrame.copy(df) # fazendo uma copia do dataframe com nome df2 para não mexer no df anterior
                            # A biblioteca Labelencoder(da sklearn) faz essa função
```

```python
df2['Sex'].replace({'F':0,'M':1},inplace=True)
df2['ChestPainType'].replace({'ATA':0,'NAP':1,'ASY':2,'TA':3},inplace=True)
df2['RestingECG'].replace({'Normal':0,'ST':1,'LVH':2},inplace=True)
df2['ExerciseAngina'].replace({'N':0,'Y':1},inplace=True)
df2['ST_Slope'].replace({'Up':0,'Flat':1,'Down':2},inplace=True)
```

**Análise das escalas dos atributos (Escalonamento)**

O escalonamento em análise de dados refere-se a técnicas utilizadas para transformar ou ajustar os valores de dados de um conjunto para que estejam dentro de um intervalo ou escala específica

```python
df.describe()
```

Padronização (utiliza a média e o desvio padrão como referência)

Normalização (utiliza os valores máximos e minimos como referência)



```python
from sklearn.preprocessing import StandardScaler # StandardScaler  = escala de padronização
                                                 # transforma os atributos de um conjunto de dados para que eles tenham média 0 e desvio padrão 1
```

```python
previsores_esc = StandardScaler().fit_transform (previsores) # transformação em previsores com relação a sua escala,
                                                             # transforma os valore proximo de zero e o desvio padão proximo de Um
```

```python
previsoresdf = pd.DataFrame(previsores_esc) # tansformar os valores escalonaveis numericos em dataframe
```

# **Base de treino e teste**

```python
from sklearn.model_selection import train_test_split
```

```python
# x_treino, x_teste são os previsores
# y_treino,y_teste são o alvo 'target'
# Arrays : atributos previsores e alvo
# train_size: tamanho em porcentagem dos dados de treinamento
# test_size : tamanho em porcentagem dos dados de teste, ou seja, separei 30% para teste
# random_state: nomeação de um estado aleatório

x_treino, x_teste, y_treino, y_teste = train_test_split (previsores_esc,alvo,test_size=0.3,random_state=0)
```

```python
x_treino.shape # 641 registros para x_treino
```

```python
x_teste.shape
```

```python
y_treino.shape
```

```python
y_teste.shape
```

# **Regressão Logística**

```python
from sklearn.linear_model import LogisticRegression # Regressão Logística: Modelo linear para classificação.
```

```python
logistica = LogisticRegression(random_state=1,max_iter=200,penalty='l2',
                                tol=0.0001,C=1,solver='lbfgs')
logistica.fit(x_treino,y_treino)
```

```python
previsao_logistica = logistica.predict(x_teste)  # são dados de previsao que iremos comparar com y_teste
previsao_logistica
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report # carregar metricas
```

```python
#accuracy_score(y_teste, previsao_logistica): Essa função compara as previsões do modelo (previsao_logistica)
#com os valores reais (y_teste) e calcula a proporção de acertos
print('Acurácia: %.2f%%' % (accuracy_score(y_teste,previsao_logistica)* 100))
```

## **Matriz de Confusão**
